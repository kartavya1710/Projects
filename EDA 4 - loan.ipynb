```python
mydata["Gender"]=mydata["Gender"].replace(np.nan,modea)
mydata["Married"]=mydata["Married"].replace(np.nan,modee)
mydata["Dependents"]=mydata["Dependents"].replace(np.nan,modeb)
mydata["Self_Employed"]=mydata["Self_Employed"].replace(np.nan,modec)

```

```python
med1 = mydata["LoanAmount"].median()
med2 = mydata["Loan_Amount_Term"].median()
med3 = mydata["Credit_History"].median()

mydata["LoanAmount"].replace(np.nan,med1,inplace=True)
mydata["Loan_Amount_Term"].replace(np.nan,med2,inplace=True)
mydata["Credit_History"].replace(np.nan,med3,inplace=True)
```
